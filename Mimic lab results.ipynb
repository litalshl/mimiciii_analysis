```python
# Create a database connection
user = 'postgres'
password = 'master'
host = 'localhost'
dbname = 'mimic'
schema = 'mimiciii'
```

```python
# Connect to the database
con = psycopg2.connect(dbname=dbname, user=user, host=host, 
                       password=password)
cur = con.cursor()
cur.execute('SET search_path to {}'.format(schema))
```

```python
# Get length of stay from the icustays table
query = \
"""
select diagnosis, l.subject_id, label, value, valueuom 
from mimiciii.labevents l inner join d_labitems dl on
l.itemid = dl.itemid inner join admissions a on
a.subject_id = l.subject_id
limit 100
"""

data = pd.read_sql_query(query,con)
```
